```python
%pip install seaborn

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('college_placement_dataset_200.csv')
```

```python
df.head()
```

```python
x = df.iloc[:,:-1]
```

```python
y = df['placed']
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
%pip install scikit-learn

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(x_train, y_train)
```

```python
model.score(x_test, y_test)*100
```

```python
from sklearn.metrics import confusion_matrix,precision_score,recall_score,f1_score 
```

```python
cf = confusion_matrix(y_test, model.predict(x_test))
cf
```

```python
sns.heatmap(cf, annot = True)
plt.show()
```

```python
 precision_score(y_test, model.predict(x_test))*100
```

```python
f1_score(y_test, model.predict(x_test))*100
```

```python
 recall_score(y_test, model.predict(x_test))*100
```
