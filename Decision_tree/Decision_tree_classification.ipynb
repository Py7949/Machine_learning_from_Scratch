```python
df = pd.read_csv('D:\Machine learning\Decision_tree\Social_Network_Ads.csv')
```

```python
sns.scatterplot(x='Age', y='EstimatedSalary', hue='Purchased', data=df)
plt.show()
```

```python
df.isnull().sum()
```

```python
x = df.iloc[:, :-1]
y = df.iloc[:, -1]
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
sc.fit(x)
x = pd.DataFrame(sc.transform(x),columns=x.columns)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
dt.fit(x_train, y_train)
```

```python
dt.score(x_test, y_test)*100
```

```python
from sklearn.tree import DecisionTreeClassifier
dt1 = DecisionTreeClassifier(criterion='entropy')
dt1.fit(x_train, y_train)
```

```python
dt1.score(x_test, y_test)*100
```

```python
from sklearn.tree import plot_tree
plt.figure(figsize=(20, 10))
plot_tree(dt)
plt.savefig('demo.jpg')
plt.show()
```

```python
# Make sure to use only the first two features for 2D decision region plotting
plot_decision_regions(x_train.iloc[:, :2].to_numpy(), y_train.to_numpy(), clf=dt)
plt.show()
```
