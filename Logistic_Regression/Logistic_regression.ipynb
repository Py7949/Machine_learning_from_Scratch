```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv(r'D:\Machine learning\Logistic_Regression\Social_Network_Ads.csv')
df.drop(['User ID','Gender','EstimatedSalary'], axis=1, inplace=True)
```

```python
df.head()
```

```python
plt.Figure(figsize=(4,3))
sns.scatterplot(x='Age', y='Purchased', data=df )
plt.show()
```

```python
x = df[['Age']]
y = df['Purchased']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(x_train, y_train)
```

```python
lr.score(x_test, y_test)*100
```

```python
lr.predict([[40]])
```

```python
plt.Figure(figsize=(4,3))
sns.scatterplot(x='Age', y='Purchased', data=df, hue='Purchased')
sns.lineplot(x=x_test['Age'], y=lr.predict_proba(x_test)[:, 1], color='red', label='Logistic Regression')
plt.legend()
plt.show()
```
