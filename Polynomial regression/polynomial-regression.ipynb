```python

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python

df = pd.read_csv(r'D:\Machine learning\Position_Salaries.csv')
df.head(3)
```

```python
x=df.iloc[:,1:-1]
```

```python
y = df.iloc[:,-1]
```

```python
sns.scatterplot(data = df,x='Level',y='Salary')
plt.title('Position vs Salary')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```

```python
df.select_dtypes(include=[np.number]).corr()
```

```python
x = df[['Level']]
y = df[['Salary']]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y,train_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train, y_train)
```

```python
from sklearn.preprocessing import PolynomialFeatures
pf = PolynomialFeatures(degree=4)
poly = pf.fit_transform(x)
lin_reg_2 = LinearRegression()
lin_reg_2.fit(poly, y)
```

```python
prd = lr.predict(x)
lr.score(x_test, y_test)*100
```

```python
sns.scatterplot(data = df, x='Level', y='Salary')
plt.plot(df['Level'], prd, color='red')
plt.title('Position vs Salary')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```

```python
sns.scatterplot(data = df, x='Level', y='Salary')
plt.plot(df['Level'], lin_reg_2.predict(pf.fit_transform(x)), color='red')
plt.title('Position vs Salary')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```

```python
x_grid = np.arange(x['Level'].min(), x['Level'].max(), 0.1)
x_grid = x_grid.reshape((len(x_grid), 1))
plt.scatter(x, y, color = 'red')
plt.plot(x_grid, lin_reg_2.predict(pf.fit_transform(x_grid)), color = 'blue')
plt.title('Truth or Bluff (Polynomial Regression)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```
