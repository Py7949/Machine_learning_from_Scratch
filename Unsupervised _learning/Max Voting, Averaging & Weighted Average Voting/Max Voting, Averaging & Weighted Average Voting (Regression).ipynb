```python
import pandas as pd
```

```python
df = pd.read_csv('D:\Machine learning\placement.csv')
```

```python
df.head()
```

```python
x = df.iloc[:, :-1]
y = df.iloc[:, -1]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
```

```python
lr = LinearRegression()
lr.fit(x_train, y_train)
```

```python
print("LinearRegression Score - Test: {:.2f} %, Train: {:.2f}%".format(lr.score(x_test, y_test)*100, lr.score(x_train, y_train)*100))

```

```python
dt = DecisionTreeRegressor()
dt.fit(x_train, y_train)
```

```python
print("Decision Tree Regressor - Test: {:.2f} %, Train: {:.2f}%".format(dt.score(x_test, y_test)*100, dt.score(x_train, y_train)*100))
```

```python
sv = SVR()
sv.fit(x_train, y_train)
```

```python
print("SVM - Test: {:.2f} %, Train: {:.2f}%".format(sv.score(x_test, y_test)*100, sv.score(x_train, y_train)*100))
```

```python
from sklearn.ensemble import VotingRegressor
```

```python
vc = VotingRegressor(estimators=[('lr', lr), ('dt', dt), ('sv', sv)])
vc.fit(x_train, y_train)
```

```python
print("VotingRegressor Accuracy {:.2f}".format(vc.score(x_test, y_test) * 100))
print("VotingRegressor Accuracy {:.2f}".format(vc.score(x_train, y_train) * 100))
```

```python
y_pred = {'LinearRegression': lr.predict(x_test),
          'DecisionTreeRegressor': dt.predict(x_test),
          'SVR': sv.predict(x_test),
          'VotingRegressor': vc.predict(x_test)}
```
